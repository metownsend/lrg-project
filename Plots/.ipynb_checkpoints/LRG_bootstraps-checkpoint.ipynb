### Interpretation

The distribution of the bootstrapped median of all LRGs in the sample is evidently bimodal. There is a peak round 9 and a peak around 10 with a significant dip in between. The median number of satellites is the same as the median found from bootstrapping: 9.84586904539.

As becomes evident from the histograms of the boostrapped median number of satellite galaxies, the bimodality starts to appear a little earlier in the the redshift slices as compared to the magnitude slices. It's possible the bimodality seen in the histogram for every LRG is driven by an evolution in the number of satellite galaxies with redshift. This is supported by plots of median number of satellites vs redshift bin and vs magnitude bin in LRG_Project_Plots.ipynb. Those plots show almost no evolution of the number of satellites in magnitude (though that gets a little iffy with gmag), while the change with redshift is obvious in the redshift plot.
